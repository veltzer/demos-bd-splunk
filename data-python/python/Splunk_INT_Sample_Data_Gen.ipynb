```python
#linux_secure
import json
import random
from faker import Faker

# Initialize Faker to generate random data
fake = Faker()

# Common values for fields
facilities = ["auth", "authpriv", "cron", "kern", "syslog"]
severities = ["info", "notice", "warning", "error"]
authentication_status = ["success", "failure"]
session_status = ["opened", "closed"]
vendor_actions = ["accepted", "failed password", "session opened"]

# Number of logs to generate
num_logs = 5000

# List to store generated logs
logs = []

# Generate logs
for _ in range(num_logs):
    log = {
        "host": fake.hostname(),
        "source": "/var/log/secure",
        "sourcetype": "linux_secure",
        "timestamp": fake.date_time_this_year().isoformat(),
        "message": fake.sentence(),
        "facility": random.choice(facilities),
        "severity": random.choice(severities),
        "pid": random.randint(1000, 9999),
        "program": fake.word(),
        "user": fake.user_name(),
        "uid": random.randint(1000, 9999),
        "ip": fake.ipv4(),
        "port": random.randint(1024, 65535),
        "authentication": random.choice(authentication_status),
        "session": random.choice(session_status),
        "event_code": random.randint(100, 999),
        "vendor_action": random.choice(vendor_actions),
    }
    logs.append(log)

# Write logs to a JSON file
output_filename = "sample_linux_secure_data.json"
with open(output_filename, "w") as f:
    json.dump(logs, f, indent=2)

print(f"Generated {num_logs} Linux secure logs with all fields in lowercase and saved to {output_filename}")
```

```python
#cisco_firewall
import json
import random
from faker import Faker

# Initialize Faker to generate random data
fake = Faker()

# Possible values for fields
dept_values = ["HR", "IT", "Finance", "Marketing", "Operations"]
location_values = ["New York", "Los Angeles", "Chicago", "Houston", "Miami"]
splunk_role_values = ["Admin", "User", "Operator"]

# Number of logs to generate
num_logs = 5000

# List to store generated logs
logs = []

# Generate logs
for _ in range(num_logs):
    log = {
        "timestamp": fake.date_time_this_year().isoformat(),
        "bcg_ip": fake.ipv4(),
        "dept": random.choice(dept_values),
        "Duration": fake.random_int(1, 100),
        "fname": fake.first_name(),
        "IP": fake.ipv4(),
        "lname": fake.last_name(),
        "location": random.choice(location_values),
        "rfid": fake.uuid4(),
        "splunk_role": random.choice(splunk_role_values),
        "splunk_server": fake.hostname(),
        "Username": fake.user_name(),
    }
    logs.append(log)

# Write logs to a JSON file
output_filename = "sample_cisco_firewall_data.json"
with open(output_filename, "w") as f:
    json.dump(logs, f, indent=2)

print(f"Generated {num_logs} Cisco firewall entries with custom sourcetype and saved to {output_filename}")
```

```python
#simecube data
import random
from faker import Faker

# Initialize Faker to generate random data
fake = Faker()

# Number of logs to generate
num_logs = 5000

# List to store generated logs
logs = []

# Generate logs
for _ in range(num_logs):
    timestamp = fake.date_time_this_year().strftime("%Y-%m-%dT%H:%M:%S.%f%z")
    src = fake.ipv4()
    version = fake.random_element(["1.0.0", "2.1.3", "3.2.1", "4.0.0"])
    misc = fake.word()
    
    # Create the composed field with user, CharacterName, action, and role
    user = fake.email()
    character_name = fake.first_name()
    action = random.choice(["attacked", "defended", "collected", "discovered"])
    role = random.choice(["Office", "Warehouse", "Sales", "Support"])
    
    composed_field = f"user:{user} CharacterName:{character_name} action:{action} role:{role}"
    
    log = [
        timestamp,
        src,
        version,
        misc,
        composed_field
    ]
    logs.append(",".join(map(str, log)))

# Write logs to a text file
output_filename = "sample_sim_cube_beta_data.txt"
with open(output_filename, "w") as f:
    f.write("\n".join(logs))

print(f"Generated {num_logs} SimCubeBeta entries with custom sourcetype and saved to {output_filename}")
```
